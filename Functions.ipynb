## Q.9 Explain the purpose and usage of the `map()` function in Python.
  - > The **map()** function is used to apply a function to each item in an iterable.
  - > **syntax:-** map(function, iterable)
  - > **usage** :-

```python
nums = [1, 2, 3, 4, 5]

squares = map(lambda x: x**2, nums)
print(list(squares))
```

## Q.10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?
  - > **map()**
    - > Apply a function to every element in an iterable.
    - > Returns an interator of elements.
    - > **syntax:-** map(function, iterable)
    - > **example** :-
    

```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)
```

- > **filter()**
  - > It selects elements from an iterable based on **condition**.
  - > Returns an interator.
  - > **example** :-

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

- > **reduce()**
  - > It combines an iterable into a **single value**.
  - > To use this we have to use a module called **functools**.
  - > **example** :-

```python
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)
```

## Q.1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
```

```python
sum_even([1,2,3,4,5,6,7,8]) #using the function
```

## Q.3  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def sqr_num(x):
  result = []
  for i in range(x):
    result.append(i**2)
  return result
```

```python
sqr_num(11)
```

## Q.4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # check up to square root of n
        if n % i == 0:
            return False
    return True
```

## Q.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fib(n):
    a=0
    b=1
    for i in range(n):
        print(a)
        a, b = b, a + b
```

```python
fib(4)
```

## Q.6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power(n):
  for i in range(n):
    yield 2**i
for i in power(5):
  print(i)
```

## Q.7  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')

# Creating a test file
with open("example.txt", "w") as f:
    f.write("First line\nSecond line\nThird line")

for line in read_file("example.txt"):
    print(line)
```

## Q.10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, s))
```

```python
text = "This is PW Skills!"
result = remove_vowels(text)
print(result)
```
